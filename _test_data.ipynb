```python
import torch
```

```python
x = torch.linspace(0, 1, 200).unsqueeze(0).unsqueeze(0)
print(x, x.shape)
```

```python
dx = x.shape[-1]
noise = torch.clamp(torch.randn_like(x)*1e-7**0.5, min=0).view(-1, dx)
y = torch.exp(torch.distributions.Normal(0.5, 1).log_prob(x)) + noise
y1 = torch.exp(torch.distributions.Normal(0.5, 0.7).log_prob(x)) + noise
```

```python
from matplotlib import pyplot as plt

plt.plot(x.squeeze(), y.squeeze())
plt.plot(x.squeeze(), y1.squeeze())
fig = plt.figure()
```
