```python
batch_size = 7
cnp = CNP(input_dim=1, hidden_dim=158, output_dim=3, n_max_obs=6, n_max_tar=6, num_layers=3, batch_size=batch_size)
wta = WTA_CNP(1, 3, 6, 6, [128, 128, 128], num_decoders=2, decoder_hidden_dims=[128, 128, 128], batch_size=batch_size)

cnp.load_state_dict(torch.load(cnp_model_path))
wta.load_state_dict(torch.load(wta_model_path))

cnp.eval()
wta.eval()
```

```python
import os
import csv

def get_batch(idx=[0, -1], cond_idx=[0]):

    batch_obs = torch.zeros(batch_size, len(cond_idx), dy+1)
    batch_obs[:, :, 0] = x[0, cond_idx, 0]

    raw_data = []
    for demo_file in sorted(os.listdir(raw_data_path)):
        if demo_file.endswith('.csv'):
            d = os.path.join(raw_data_path, demo_file)

            temp_data = []
            
            with open(d, 'r') as f:
                for i, line in enumerate(csv.reader(f)):
                    if i > 0:
                        temp_data.append([float(line[3]), float(line[4]), float(line[5])])  # x, y, z

            raw_data.append(temp_data)

    demonstration = []
    for i in range(len(raw_data)):
        traj = raw_data[i]
        processed_traj = []
        inds = torch.linspace(0, len(traj)-1, t_steps, dtype=int)
        for ind in inds:
            processed_traj.append(traj[ind])

        demonstration.append(processed_traj)

    batch_y = torch.tensor(demonstration, dtype=torch.float32)

    batch_obs[:, :, 1:] = batch_y[idx, cond_idx, :].view(batch_size, len(cond_idx), dy)
    # for i in range(batch_size):
    #     batch_obs[i, 0, 0] = x[i, cond_idx[0], 0]
    #     batch_obs[i, 0, 1:] = batch_y[i, cond_idx[0], :]
    return batch_obs, batch_y[idx, :, :]


```

```python
import numpy as np

# load the demonstration file, change the values, interpolate as necessary, save it 
import csv
import scipy.interpolate as interp


demo_file = f'{raw_data_path}used/after_pick_2023-11-16_18-30-05.csv'
demo_data = []

#read into demo_data
with open(demo_file, 'r') as f:
    csv_reader = csv.reader(f)
    
    for i, line in enumerate(csv_reader):
        if i > 0:
            line_flt = [float(x) for x in line]
            demo_data.append(np.array(line_flt))
        else:
            header = line

demo_data = np.array(demo_data)
print(demo_data.shape)


def p_control(cond, traj, decay_steps=10, kp=0.5):
    decay_factor = 1 / decay_steps
    corrections = np.zeros_like(traj)
    error = traj[0] - cond
    initial_correction = -kp * error

    # Gradually decrease the correction values to 0 in the specified number of steps
    for i in range(decay_steps):
        corrections[i] = initial_correction * (1 - i * decay_factor)
    modified_traj = np.insert(traj + corrections, 0 , cond, axis=0)
    return modified_traj[:-1]

    # modified_traj = traj + corrections

    # return modified_traj


def save_traj(traj, file_name):
    with open(file_name, mode='w') as outfile:
        writer = csv.writer(outfile)
        writer.writerow(header)
        for i in range(len(traj)):
            writer.writerow(traj[i])


def process_trajs(index):
    best = torch.argmax(gate[index, 0, :]).item()

    wta_traj = pred_wta[best, index, :, :dy].cpu().numpy()
    cnp_traj = pred_cnp[index, :, :dy].cpu().numpy()

    gen_path = f'{root_path}{run_id}/gen/'
    if not os.path.exists(f'{gen_path}'):
        os.makedirs(f'{gen_path}')

    wta_gen_file = f'{gen_path}cnep_{index}_xyz.csv'
    cnp_gen_file = f'{gen_path}cnmp_{index}_xyz.csv'

    # interpolate generated trajectories wrt first column (time)
    start, end = demo_data[0][0], demo_data[-1][0]
    len_traj = len(wta_traj)
    x = np.linspace(start, end, len_traj)

    fx = interp.interp1d(x, wta_traj[:, 0])
    fy = interp.interp1d(x, wta_traj[:, 1])
    fz = interp.interp1d(x, wta_traj[:, 2])
    wta_interp, cnp_interp = np.zeros((len(demo_data[:, 0]), dy)), np.zeros((len(demo_data[:, 0]), dy))
    wta_interp[:, 0] = fx(demo_data[:, 0])
    wta_interp[:, 1] = fy(demo_data[:, 0])
    wta_interp[:, 2] = fz(demo_data[:, 0])

    fx = interp.interp1d(x, cnp_traj[:, 0])
    fy = interp.interp1d(x, cnp_traj[:, 1])
    fz = interp.interp1d(x, cnp_traj[:, 2])
    
    cnp_interp[:, 0] = fx(demo_data[:, 0])
    cnp_interp[:, 1] = fy(demo_data[:, 0])
    cnp_interp[:, 2] = fz(demo_data[:, 0])

    # p-control
    wta_p_control = p_control(obs[index, :, 1:4].cpu().numpy(), wta_interp)  # 1:4 because 1:x, 2:y, 3:z
    cnp_p_control = p_control(obs[index, :, 1:4].cpu().numpy(), cnp_interp)

    # replace columns 3, 4 & 5 (x,y,z) of demo_data with learned, interpolated and p-controlled values
    demo_data[:, 3:3+dy] = wta_p_control
    save_traj(demo_data, wta_gen_file)

    demo_data[:, 3:3+dy] = cnp_p_control
    save_traj(demo_data, cnp_gen_file)

    return wta_p_control, cnp_p_control
```

```python
rows, cols = torch.floor(torch.sqrt(torch.tensor(batch_size))).int(), torch.ceil(torch.sqrt(torch.tensor(batch_size))).int()
ax_rows, ax_cols = rows.item(), cols.item()  
fig_size_coeff = 8
fig, axs = plt.subplots(ax_rows, ax_cols, figsize=(ax_cols*fig_size_coeff, ax_rows*fig_size_coeff))
for i in range(ax_rows):
    for j in range(ax_cols):
        ind = i*ax_cols + j

        subplot_index = determine_subplot_index(i,j,ax_rows)
        # axs[subplot_index].set_xlim((0.55, 1.15))
        # axs[subplot_index].set_ylim((-0.55, -0.05))

        axs[subplot_index].scatter(obs[ind, :, 1], obs[ind, :, 2], label='Observation', c='k', s=250.0)

        w, c = process_trajs(ind)
        
        axs[subplot_index].plot(c[:, 0], c[:, 1], color=palette[0], linewidth=4.0)  # cnp predictions

        handles = axs[subplot_index].get_legend_handles_labels()[0]
        handles.extend([Line2D([0], [0], label='CNMP', color=palette[0])])


        axs[subplot_index].plot(w[:, 0], w[:, 1], color=palette[1], linewidth=4.0)  # wta predictions
        handles.extend([Line2D([0], [0], label=f'CNEP', color=palette[1])])

        axs[subplot_index].plot(y[0,:,0], y[0,:,1], 'k:', alpha=0.85, linewidth=3.0)  # data
        axs[subplot_index].plot(y[1,:,0], y[1,:,1], 'k:', alpha=0.85, linewidth=3.0)  # data
        handles.extend([Line2D([0], [0], label='Demonstrations', alpha=0.75, color='k', linestyle='dotted')])

        axs[subplot_index].legend(handles=handles, loc='lower left')
        axs[subplot_index].set_title(f'Test: {ind+1}')
        axs[subplot_index].grid(True)
        axs[subplot_index].set_xlabel('x')
        axs[subplot_index].set_ylabel('y')

fig.suptitle(f'Generated Trajectories (Trained on the Cartesian Space)', fontsize=24)
plt.tight_layout(rect=[0, 0.03, 1, 0.95])
# plt.savefig(f'{root_path}{run_id}/traj_generation.png')
plt.show()
```
